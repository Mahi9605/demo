Pie Plot

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y= ["c","c++","java","python"]
ex= [0.2,0.0,0.0,0.0]
                                                                                                                                                                                                             
plt.pie(x,labels=y,explode=ex,autopct= "%0.1f%%",radius=1,startangle=0)
plt.title("Pie Plot")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[30,40,50,20]
y= ["c","c++","java","python"]

c=["r","m","b","y"]

plt.pie(x, labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
x1=[30,40,50,20]
y= ["c","c++","java","python"]

c=["r","m","b","y"]

plt.pie(x, labels=y,radius=1.5)
cr = plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)
plt.show()
```
