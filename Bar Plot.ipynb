```python
import matplotlib.pyplot as plt
x=["python", "c", "c++", "java"]
y=[89,60,75,85]
plt.xlabel("Langugage", fontsize =15)
plt.ylabel("No")
c=["y","b","m","r"]
plt.bar(x,y,width=0.5,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python", "c", "c++", "java"]
y=[89,60,75,85]
z=[10,20,30,40]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Langugage", fontsize =15)
plt.ylabel("No",fontsize=15)

plt.bar(p,y,width,color="r")
plt.bar(p1,z, width,color="y")
plt.xticks(p+width/2,x)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python", "c", "c++", "java"]
y=[89,60,75,85]
z=[10,20,30,40]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Langugage", fontsize =15)
plt.ylabel("No",fontsize=15)

plt.barh(p,y,width,color="r")
plt.barh(p1,z, width,color="y")
plt.xticks(p+width/2,x)
plt.show()
```
